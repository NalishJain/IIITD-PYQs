2.
- The first plot demonstrates the original signal's behavior with varying α values.\
- The second plot showcases the even component.\
- The third plot depicts the odd component.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-100, 100, 1000)

alpha = float(input())
x = 2 * np.exp(-alpha * t)
x[t < 0] = 0
x_even = 0.5 * (x + np.flip(x))
x_odd = 0.5 * (x - np.flip(x))

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, x, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Original Signal x(t) = 2e^{-\\alpha t}')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, x_even)
plt.xlabel('t')
plt.ylabel('Even Component')
plt.title('Even Component of x(t)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, x_odd)
plt.xlabel('t')
plt.ylabel('Odd Component')
plt.title('Odd Component of x(t)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
