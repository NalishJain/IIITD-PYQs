```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_series(N, N1):
    a = np.zeros(N, dtype=complex)
    for k in range(0, N):
        for n in range(-N1, N1 + 1):
            a[k] += (1 / N) * np.exp(-1j * (2 * np.pi / N) * k * n)
    return a

N1 = 2

N_values = [4 * N1 + 1, 8 * N1 + 1, 10 * N1 + 1]
n_values = np.arange(-N1, N1 + 1)
plt.figure(figsize=(8, 4))
for N in N_values:
    coefficients = fourier_series(N, N1)
    print(f"For N = {N}:")
    for k, Xk in enumerate(coefficients):
        print(f"X[{k}] = {Xk}")
        plt.stem(k, Xk)
        print(k)
        print(Xk)
    plt.xlabel('n')
    plt.ylabel('x[n]')
    plt.title(f'Fourier Series for N = {N}')
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x + np.pi

x = np.linspace(-np.pi, np.pi, 1000)

a_0 = 0
a_n = [2 * (-1) ** n / n for n in range(1, 100)]
b_n = [2 * (1 - (-1) ** n) / (n * np.pi) for n in range(1, 100)]

def fourier_series(x, a_0, a_n, b_n, n_terms):
    result = a_0 / 2
    for n in range(n_terms):
        result += a_n[n] * np.cos(n * x) + b_n[n] * np.sin(n * x)
    return result

n_terms = 10
y_series = fourier_series(x, a_0, a_n, b_n, n_terms)

plt.figure(figsize=(10, 6))
plt.stem(x, y_series)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
